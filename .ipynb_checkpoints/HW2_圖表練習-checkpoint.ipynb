```python
import pandas as pd
df = pd.read_csv("recent_grads.csv")
df
#人工智慧科技基金會線上教學(https://www.youtube.com/watch?v=BbC3cAm4V78)
#運用{大學現況}的資料，原作者有更新過
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.info()
```

```python
df.plot(x="Rank", y=["P25th","Median","P75th"])
plt.show()
#科系收入跟收入的變化
#折線圖(x=數值 y=數值)
```

```python
df.plot(x="Median", y = "Unemployment_rate", kind = "scatter")
plt.show()
#系所收入中位數與失業人數比例關係
#散佈圖(x=數值 y=數值)
```

```python
df.plot(x="Rank", y="Median", kind='scatter')
df.plot(x="Rank", y="P25th", kind='scatter')
plt.show()
#散佈圖沒辦法疊加
```

```python
top_5 = df.sort_values(by="Median",ascending=False).head()
#sort_values重新計算，False大排到小、True小排到大
top_5.plot(x="Major",y="Median", kind="bar", fontsize=15)
plt.show()
#收入中位數前五名的科系在收入中位數上的差異
#長條圖
```

```python
top_medians = df[df["Median"]> 70000].sort_values("Median")
top_medians.plot(x="Major",y=["P25th","Median","P75th"], kind="bar")
plt.show()
#收入中位數超果過7萬的科系在收入上的差異
#長條圖2
```

```python
cat_medians = df.groupby("Major_category")["Median"].mean().sort_values()
cat_medians.plot(kind="barh", fontsize=15)
#bar是直的，加h就變水平
plt.show()
#不同科系的類群的中位數差異 
#長條圖3
```

```python
median_column = df["Median"]
median_column.plot(kind="hist")
plt.show()
#全部科系的收入中位數分布範圍跟集中程度
#直方圖(X=數值，Y=數值(次數))
#統計分析
```
